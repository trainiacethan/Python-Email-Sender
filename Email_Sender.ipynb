```python
import smtplib
from email.message import EmailMessage

# Define configuration variables
SMTP_SERVER = "smtp.gmail.com"  # Replace with your provider's SMTP server
SMTP_PORT = 465                 # Port 465 is used for secure SSL connections
SENDER_EMAIL = "example@email.com"
APP_PASSWORD = ""               # https://myaccount.google.com/apppasswords - create a new app password, enter here
RECEIVER_EMAIL = "example@email.com"

# Build the email headers and body
msg = EmailMessage()
msg["Subject"] = "Automated Python Email Test"
msg["From"] = SENDER_EMAIL
msg["To"] = RECEIVER_EMAIL
msg.set_content("Hello! This email was sent successfully using Python.")

# Establish a secure connection and send the email
try:
    with smtplib.SMTP_SSL(SMTP_SERVER, SMTP_PORT) as server:
        server.login(SENDER_EMAIL, APP_PASSWORD)
        server.send_message(msg)
    print("Email sent successfully!")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import smtplib
from email.message import EmailMessage
import mimetypes

msg = EmailMessage()
msg["Subject"] = "Another Email"
msg["From"] = "example@email.com"
msg["To"] = "example@email.com"

msg.set_content("Please enable HTML to view this message.")

# HTML Content Layer
html_content = """
<html>
    <body>
        <h2 style="color: #2e6c9e;">yo whasgood</h2>
        <p>This is the <strong>Ferrari 488 GTE</strong> and a <a href="https://www.pagani.com/huayra-bc/">link</a>!</p>
    </body>
</html>
"""
msg.add_alternative(html_content, subtype="html")

# Add a file
file_path = "488GTE.jpg"
with open(file_path, "rb") as f:
    file_data = f.read()
    file_name = f.name

# Automatically detect the file type (PDF, PNG, etc.)
mime_type, _ = mimetypes.guess_type(file_path)
main_type, sub_type = mime_type.split("/", 1)

msg.add_attachment(file_data, maintype=main_type, subtype=sub_type, filename=file_name)

# Send via SMTP_SSL
with smtplib.SMTP_SSL("smtp.gmail.com", 465) as server:
    server.login("example@email.com", APP_PASSWORD)
    server.send_message(msg)
```
